# Question 3 : 

Calculez (pour chaque pays et chaque produit) la disponibilité alimentaire en kcal puis en kg de protéines

Vous ferez cela à partir de ces informations : Population de chaque pays ; Disponibilité alimentaire ("Food Supply" en anglais) donnée pour chaque produit et pour chaque pays en kcal/personne/jour. Disponibilité alimentaire en protéines ("Protein supply quantity" en anglais) donnée pour chaque produit et pour chaque pays en g/personne/jour.

```python
# importation de la table bilan_2013
bilan_2013 = pd.read_csv("bilan_2013.csv")
```

```python
# formatage du dataframe principal
bilan_2013 = bilan_2013.pivot_table(values = ["Valeur"], index = ["Code Pays", "Pays", "Code Produit", "Produit", 
                                                                  "Année"],columns = ["Élément"], aggfunc = sum)
bilan_2013.columns = ['Aliments pour animaux', 'Autres Utilisations', 
                      'Disponibilité alimentaire (Kcal/personne/jour)', 
                      'Disponibilité alimentaire en quantité (kg/personne/an)', 
                      'Disponibilité de matière grasse en quantité (g/personne/jour)', 
                      'Disponibilité de protéines en quantité (g/personne/jour)', 'Disponibilité intérieure', 
                      'Exportations - Quantité', 'Importations - Quantité', 'Nourriture', 'Pertes', 
                      'Population totale','Production', 'Semences', 'Traitement', 'Variation de stock']
bilan_2013 = bilan_2013.reset_index()
bilan_2013.head()
```

```python
# mise à l'unité utile des variables suivantes :
bilan_2013['Aliments pour animaux'] = bilan_2013['Aliments pour animaux']*1000000
bilan_2013['Autres Utilisations'] = bilan_2013['Autres Utilisations']*1000000
bilan_2013['Importations - Quantité'] = bilan_2013['Importations - Quantité']*1000000
bilan_2013['Exportations - Quantité'] = bilan_2013['Exportations - Quantité']*1000000
bilan_2013['Pertes'] = bilan_2013['Pertes']*1000000
bilan_2013['Production'] = bilan_2013['Production']*1000000
bilan_2013['Semences'] = bilan_2013['Semences']*1000000
bilan_2013['Traitement'] = bilan_2013['Traitement']*1000000
bilan_2013['Variation de stock'] = bilan_2013['Variation de stock']*1000000
bilan_2013['Disponibilité intérieure'] = bilan_2013['Disponibilité intérieure']*1000000
bilan_2013['Nourriture'] = bilan_2013['Nourriture']*1000000
bilan_2013['Population totale'] = bilan_2013['Population totale']*1000
```

```python
# pour remplacer les valeurs 'NaN' par une valeur numérique
bilan_2013 = bilan_2013.fillna(0)

# pour obtenir une valeur sur chacune des lignes de la variable 'Population totale'
for elt in pd.unique(bilan_2013['Pays']):
    bilan_2013.loc[bilan_2013['Pays'] == elt,
                   'Population totale'] = bilan_2013.loc[bilan_2013['Pays'] == elt,'Population totale'].max()
    
# suppression de l'individu Chine en doublon
bilan_2013 = bilan_2013.loc[bilan_2013.Pays != 'Chine', :]

# pour supprimer la valeur 'Population' de l'ensemble de la variable 'Produit'
bilan_2013 = bilan_2013.loc[bilan_2013.Produit != 'Population', :]
```

```python
# Calcul (pour chaque pays et chaque produit) de la disponibilité alimentaire en kcal

# création d'une nouvelle colonne résultante de la mutliplication de la colonne 
#'Disponibilité alimentaire (Kcal/personne/jour)' avec 'Population totale', le tout multiplié par le 
# nombre de jours, que compte une année
bilan_2013.loc[:, 'Dispo alim kcal (kcal/pays/an)'] = bilan_2013['Disponibilité alimentaire (Kcal/personne/jour)'] * bilan_2013['Population totale'] * 365
```

```python
# Calcul (pour chaque pays et chaque produit) de la disponibilité alimentaire en kg de protéines

# création d'une nouvelle colonne résultante de la mutliplication de la colonne 
# 'Disponibilité de protéines en quantité (g/personne/jour)' avec'Population totale', le tout divisé par 
# 1000 (pour être à l'unité kg) et multiplié par le nombre de jours, que compte une année
a = bilan_2013['Disponibilité de protéines en quantité (g/personne/jour)'] * 365 / 1000
bilan_2013.loc[:, 'Dispo alim prot (kg/pays/an)'] = a * bilan_2013['Population totale']
bilan_2013.head()
```

# Question 4 : 
A partir de ces dernières informations, et à partir du poids de la disponibilité alimentaire (pour chaque pays et chaque produit), calculez pour chaque produit le ratio "énergie/poids", que vous donnerez en kcal/kg. Vous pouvez vérifier la cohérence de votre calcul en comparant ce ratio aux données disponibles sur internet, par exemple en cherchant la valeur calorique d'un oeuf.

Indication : La disponibilité alimentaire en kcal/personne/jour est calculée par la FAO en multipliant la quantité Nouriture (Food) par le ratio énergie/poids (en kcal/kg), puis en le divisant par la population du pays puis par 365. Ici, on vous demande juste de retrouver le ratio énergie/poids que la FAO a utilisé dans son calcul.

```python
# Ratio "énergie/poids" (kcal/kg) = Disponibilité alimentaire (Kcal/personne/jour)*pop*365/Nourriture(en kg) 
# Ratio "énergie/poids" (kcal/kg) = Dispo alim (Kcal/pays/an)/Nourriture(en kg)

# création d'une nouvelle colonne résultante de l'égalité donnée ci-dessus
bilan_2013.loc[:, 'Ratio "énergie/poids" (kcal/kg)'] = bilan_2013['Dispo alim kcal (kcal/pays/an)']/bilan_2013['Nourriture']
bilan_2013.head()
```

```python
# à l'aide d'un nouveau dataframe sélection pour la variable 'Produit' de la valeur 'Oeufs'
bilan_2013_oeuf = bilan_2013[bilan_2013['Produit'] == "Oeufs"].copy()
bilan_2013_oeuf['Ratio "énergie/poids" (kcal/kg)'].mean()
```

```python
# pour remplacer les valeurs 'NaN' par une valeur numérique
bilan_2013 = bilan_2013.fillna(0)

# les valeurs 'inf' de la variable 'Ratio "énergie/poids" (kcal/kg)' sont remplacées par la valeur '0'
bilan_2013.loc[np.isinf(bilan_2013['Ratio "énergie/poids" (kcal/kg)']),'Ratio "énergie/poids" (kcal/kg)'] = 0

# on reprend la case [19]
bilan_2013_oeuf = bilan_2013[bilan_2013['Produit'] == "Oeufs"].copy()

# calcul de la moyenne sur l'ensemble des pays de la valeur 'Oeufs' au sein de la variable 
# 'Ratio "énergie/poids" (kcal/kg)'
# pour le calcul de la moyenne ne sont pris en compte, que les valeurs >0
bilan_2013_oeuf.loc[bilan_2013_oeuf['Ratio "énergie/poids" (kcal/kg)']>0,'Ratio "énergie/poids" (kcal/kg)'].mean() 
```

```python
# après vérification sur la fiche Wikipédia de l'oeuf celui-ci à une valeur calorique de 1470 kcal/kg
# notre calcul nous donne 1345 kcal/kg
```

```python
# création d'un nouveau dataframe où les lignes égales à 0 dans la variable 'Ratio kcal/kg' sont supprimées
bilan_2013_cal = bilan_2013.loc[bilan_2013['Ratio "énergie/poids" (kcal/kg)'] != 0, :]
```

```python
# tri descendant au sein du dataframe donnant les 20 aliments les plus caloriques
groupped.sort_values(by = 'moy Ratio "énergie/poids" (kcal/kg)', ascending = False)[0:20]
```

Repérez les 15 produits les plus exportés par ce groupe de pays sur l'année choisie.

Groupez ces importations par produit, afin d'avoir une table contenant 1 ligne pour chacun des 15 produits. Ensuite, calculez pour chaque produit les 2 quantités suivantes :

```python
# Afin de repérer dans nos données les informations concernant les pays dans lesquels la FAO recense des personnes 
# en sous-nutrition
# création liste 'list_sous_nut' des pays en sous-nutrition
list_sous_nut = np.unique(securite_2013.loc[securite_2013['Valeur']>0,'Code zone'])
```

```python
# création d'une nouvelle colonne affichant si le 'Code Pays' fait parti des pays donnés par la FAO ou non
bilan_2013['is_ss_nut'] = bilan_2013['Code Pays'].isin(list_sous_nut)
```

```python
bilan_2013.head()
```

```python
# création d'une nouvelle table où ne seront conservés que les pays donnés par la FAO
ss_nut_2013 = bilan_2013.loc[bilan_2013['is_ss_nut'] == 1, :]
ss_nut_2013.head()
```

```python
# pour repérer les 15 produits les plus exportés par ce groupe de pays sur 2013
# création d'un objet de type  DataFrameGroupBy, que nous appelons  gb_2 :
gb_2 = ss_nut_2013.groupby('Code Produit')
```

```python
# sur cet objet gb_2, on peut ensuite appliquer la fonction d'agrégation .sum()
s_2 = gb_2['Exportations - Quantité',].sum()

# représentation du dataframe
groupped_2 = pd.concat([s_2], axis=1)
groupped_2.columns = ["Exportations - Quantité"]
```

```python
# tri descendant au sein du dataframe 
c = groupped_2.sort_values(by = 'Exportations - Quantité', ascending = False)[0:15]
c
```

```python
# Afin de repérer dans nos données les informations concernant les produits les plus exportés par les pays en 
# sous-nutrition
# création liste 'list_prod_exp_ss_nut'
list_prod_exp_ss_nut = np.unique(c.index)
```

```python
# création d'une nouvelle colonne affichant si le 'Code Produit' fait parti de la liste ci-dessus ou non
bilan_2013['is_prod_exp_ss_nut'] = bilan_2013['Code Produit'].isin(list_prod_exp_ss_nut)
```

```python
# création d'une nouvelle table où ne seront conservées que les pays donnés ci-dessus
exp_ss_nut_2013 = bilan_2013.loc[bilan_2013['is_prod_exp_ss_nut'] == 1, :]
exp_ss_nut_2013.head()
```

```python
# tri descendant au sein du dataframe 
exp_ss_nut_2013.sort_values(by = 'Importations - Quantité', ascending = False)[0:200]
```

```python
# nous allons grouper ces importations par produit, afin d'avoir une table contenant 1 ligne pour chacun des 15 
# produits
# création d'un objet de type  DataFrameGroupBy, que nous appelons  gb_3 :
gb_3 = exp_ss_nut_2013.sort_values(by = 'Importations - Quantité', ascending = False)[0:200].groupby('Produit')
```

```python
# sur cet objet gb_3, on peut ensuite appliquer la fonction d'agrégation .sum()
s_3 = gb_3[['Importations - Quantité', 'Autres Utilisations', 'Disponibilité intérieure', 'Aliments pour animaux',
           'Nourriture']].sum()
s_3
```

```python
# calcul du ratio entre la quantité destinés aux "Autres utilisations" et la disponibilité intérieure
# création d'une nouvelle colonne résultante du ratio donné ci-dessus
s_3['Ratio Other uses/Di'] = s_3['Autres Utilisations']/s_3['Disponibilité intérieure']
```

```python
# calcul du ratio entre la quantité destinée à la nourriture animale et la quantité destinée à la nourriture
# création d'une nouvelle colonne résultante du ratio donné ci-dessus
s_3['Ratio Alim_ani/Food'] = s_3['Aliments pour animaux']/(s_3['Nourriture']+s_3['Aliments pour animaux'])
s_3
```
